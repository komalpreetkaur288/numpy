1️⃣ create array

```python
students=np.array(['Aman','Komal','Jass','Prabh','Kiran'])
```

```python
marks=np.array([55,66,84,83,38])
```

```python
#find marks for each students
for i in range(len(students)):
 print(students[i],marks[i])
```

2️⃣ find average of marks

```python
# Average of marks
average_marks=np.mean(marks)
print(average_marks)
```

3️⃣ Find minimum marks

```python
#minimum marks
min_marks=np.min(marks)
print(min_marks)
```

4️⃣ Find maximum marks

```python
#maximum marks
max_marks=np.max(marks)
print(max_marks)
```

5️⃣ Find maximum marks and also return index

```python
#maximum marks
max_marks=np.max(marks)
print(max_marks)

max_index=np.argmax(marks)
print(max_index)
print(students[max_index])
```

```python
#minimum marks
min_marks=np.min(marks)
print(min_marks)

min_index=np.argmin(marks)
print(min_index)
print(students[min_index])
```

6️⃣ pass or fail

```python
# pass or fail(marks>=40)
passed_students=marks>=40
print(passed_students )
```

7️⃣ check passing students names with score (using for loop)

```python
#check passing students names with score
for i in range(len(students)):
  if marks[i]>=40:
    print(students[i],marks[i])
```

8️⃣ check failing students names with score(using g for loop)

```python
#check failing students names with score
for i in range(len(students)):
  if marks[i]<=40:
    print(students[i],marks[i])
```

```python
median_m=np.median(marks)
print(median_m)
```
